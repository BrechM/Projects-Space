## Prétraitement des données

Le prétraitement des données consistera à gerer les valeurs manquantes, à encoder les variables catégorielles et à standardiser les variables quantitatives

Commencons d'abord par voir ce qui se passe au niveau de la variable TotalCharges.

```python
# Forcer la conversion de la variable 'TotalCharges'
# Les variables qui ne seront pas converties seront transformées en valeurs manquantes
total_charges = pd.to_numeric(train_copy['TotalCharges'], errors='coerce')
total_charges.isna().sum()
```

```python
train_copy[total_charges.isna()]
```

```python
# On a 8 valeurs qui n'ont pas etre transformées en valeur numérique
# Valeurs qui posent probleme
train_copy[total_charges.isna()]['TotalCharges'].unique()
```

array([' '], dtype=object) indique que ce sont les espaces qui causent probleme au niveau de la variables TotalCharges.

Créons maintenant un pipeline pour le traitement des données avant la phase de modélisation

```python
# Fonction pour regler le probleme de la colonne 'TotalCharges'
def convert_totalcharges(X):
    # X : dataframe
    Z = X.copy()
    Z['TotalCharges'] = pd.to_numeric(Z['TotalCharges'], errors='coerce')
    return Z.values # .values me retourne un tableau numpy
```

Le pipeline de prétraitement des données doit comporter :

- Une etape pour convertir la variable 'TotalCharges' en variable numérique;
- une etape pour gérer les valeurs manquantes de la variable 'TotalCharges';
- une etape pour standardiser les variables indépendantes (features) nimériques;
- une etape pour encoder les variables indépendantes qualitatives.

Pour commencer, créons une liste des variables indépendantes numériques et une autre liste des vraiables indépendantes qualitatives:

```python
numeric_features = quant_cols[:-1] + ['TotalCharges']
numeric_features
```

```python
categorical_features = cat_cols[:-1]
categorical_features
```

#### Construction Pipeline  

```python
# Pipeline de prétraitement des variables indépendantes numériques
numeric_transformer = Pipeline(
    steps = [('convert', FunctionTransformer(convert_totalcharges)),
             ('imputer', SimpleImputer(strategy='median')),
             ('scaler', StandardScaler())]
)

# Pipeline de prétraitement des variables indépendantes qualitatives
categorical_transformer = Pipeline(
    steps = [('OneHotEncoder', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))]
)

# Combinaison des deux précédents pipelines en un seul
preprocessor = ColumnTransformer(
    transformers=[('numeric', numeric_transformer, numeric_features),
                  ('categorical', categorical_transformer, categorical_features)]
)
print(preprocessor)
```

```python
# Image du pipeline de prétraitement (sous format Image)
preprocessor
```

## Modelisation

```python
# Construction du pipeline de modelisation avec l'algorithme des Forets aleatoires comme estimateur

rf_pipe_model = Pipeline(
    steps=[('preprocessor', preprocessor),
           ('rf', RandomForestClassifier(random_state=seed))]
)
rf_pipe_model
```

```python
# Construction du pipeline de modélisation avec l'algorithme de régression logistique comme estimateur
logreg_pipe_model = Pipeline(
    steps = [('preprocessor', preprocessor),
             ('loreg', LogisticRegression(random_state=seed, max_iter=2000))]
)

logreg_pipe_model
```

```python
X_train, X_val, y_train, y_val = train.drop('Churn', axis=1), validate.drop('Churn', axis=1), train['Churn'], validate['Churn']
print(X_train.shape, y_train.shape, X_val.shape, y_val.shape, '\n')

# .shape donne ses dimensions
# '\n' C’est un caractère de saut de ligne : il force un retour à la ligne après l’affichage des shapes.
```

```python
# Formation du modele de foret aleatoire
rf_pipe_model.fit(X_train, y_train)

# Predictions sur les données d'entrainement
y_train_preds_rf = rf_pipe_model.predict(X_train)

# Predictions sur les données de validation
y_val_preds_rf = rf_pipe_model.predict(X_val)

# Rapports de classification
print(classification_report(y_train, y_train_preds_rf))
print('\n')
print(classification_report(y_val, y_val_preds_rf))

# Courbe ROC
rf_roc = RocCurveDisplay.from_estimator(rf_pipe_model, X_val, y_val)
```

On peut soupconner qu'il y'a overfitting (car ecart entre le données de Train et les données de test)

l'AUC est de 0.78, on ne peut pas deployer en production un tel modele

```python
# Formation du modele de regression logistique
logreg_pipe_model.fit(X_train, y_train)

# Predictions sur les données d'entrainement
y_train_preds_rf = logreg_pipe_model.predict(X_train)

# Predictions sur les données de validation
y_val_preds_rf = logreg_pipe_model.predict(X_val)

# Rapports de classification
print(classification_report(y_train, y_train_preds_rf))
print('\n')
print(classification_report(y_val, y_val_preds_rf))

# Courbe ROC
logreg_roc = RocCurveDisplay.from_estimator(logreg_pipe_model, X_val, y_val)
```

D'apres les resultats ci-dessous, le modele de regression logistique est meilleur que celui de forets aleatoires.Nous allons proceder au reglage des hyperparametres de ce modele:

## Réglage des hyperparametres du modele

```python
# Reglage des hyperparamètres du modele de regression logistique

pipeline_model = Pipeline(
    steps = [('preprocessor', preprocessor),
             ('logreg', LogisticRegression(random_state=seed, max_iter=2000))]
)

param_grid = {
    'logreg__solver' : ['lbfgs', 'saga', 'sag', 'newton-cg', 'newton-cholesky'],
    'logreg__penalty': ['12', None],
    'logreg__C': [0.5, 1.0, 5.0, 10] # parametre de regularisation
}

search_model = GridSearchCV(
    estimator=pipeline_model,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=5, # fait automatiquement la validation croisée
    n_jobs=-1 # pourqu'il utilise toute la RAM de l'ordinateur
)

search_model.fit(X_train, y_train)

print(search_model.best_params_)

print(search_model.best_score_)

# On a une autre possibilite, on peut faire un random search ou un Bayesian search qui sont plus 
# optimals qu'un GridSearch 
```

```python
# Enregistrement du meilleur modele
joblib.dump(search_model.best_estimator_, 'C:/Users/brech/Desktop/dossier de taff/ML/Projet 1/models/best_model.pkl')
```
