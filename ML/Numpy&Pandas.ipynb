#### 6. Fusion et jointure de Data Frames

```python
# Creation des dataFrames exemple
data1= {
    "ID":[1, 2, 3, 4], "Prenom":["Alice", "Bob", "Charlie", "David"]
}
data2= {
    "ID":[3, 4, 5, 6], "Prenom":["Paris", "Lyon", "Marseille", "Toulouse"]
}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
# effectuer une jointure interne (inner join)
inner_join = df1.merge(df2, on="ID", how="inner")
```

```python
# effectuer une jointure externe (outer join)
outer_join = df1.merge(df2, on="ID", how="outer")
```

```python
# Effectuer une jointure gauche (left join)
left_join = df1.merge(df2, on="ID", how="left")
```

```python
# Effectuer une jointure droite (right join)
right_join = df1.merge(df2, on="ID", how="right")
```

```python
# Affichage des resultats
print("Inner_join:")
print(inner_join)
```

```python
print("\nouter join:")
print(outer_join)
```

```python
print("\nLeft join:")
print(left_join)
print("\nRight join:")
print(right_join)
```
