### Matplotlib

### 1 Installation et importation du package

```python
# Importation des packages
! pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

#### 2 Tracé de graphiques simples avec Matplotlib

```python
# Generation de données
x=np.linspace(0,10,1000)
y=np.exp(x)
```

```python
# Tracer un graphique simple
plt.plot(x,y)
# Sur d'autres environnements, on aura besoin de taper (plt.show()) pour afficher le graphique
```

### 3. Personnalisation de graphique avec Matplotlib

```python
# Changer la taille d'un graphique avec la largeur et la hauteur de la figure en pouce
plt.figure() # plt.figure(figsize=(5,5))
plt.plot(x,y)
```

```python
# Titre et axes
plt.plot(x,y)
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
plt.show()
```

```python
# Changement de couleur
plt.plot(x,y,c="red")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
plt.show()
```

```python
# Changer le type de ligne
plt.plot(x,y,c="red",ls="--")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
plt.show()
```

```python
# Ajouter une legende

plt.plot(x,y,c="red",ls="--", label="exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
plt.legend()
plt.show()
```

```python
# Ajouter une grille sur le graphique
plt.plot(x,y,c="red",ls="--", label="exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
plt.legend()
plt.grid()
plt.show()
```

### 4. Faire plusieurs graphiques

#### 4.1 Faire des graphiques sur un meme carré

```python
# Creer un graphique vide

z=2*y
y3=z*2
plt.plot(x,y,c="red",ls="-", label="exp")
plt.plot(x,z,c="blue",ls="-", label="2*exp")
plt.plot(x,y3,c="green",ls="-", label="4*exp")
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
plt.legend()
plt.grid()
plt.show()
```

```python
# Pour zoomer une partie du graphique (par exemple visionner sur (6,10))
# On va changer les limites des axes

z=2*y
y3=z*2
plt.plot(x,y,c="red",ls="-", label="exp")
plt.plot(x,z,c="blue",ls="-", label="2*exp")
plt.plot(x,y3,c="green",ls="-", label="4*exp")
plt.xlim((6,10))
plt.ylim((1000,80000))
plt.title("Fonction exponentielle")
plt.xlabel("x")
plt.ylabel("valeur")
plt.legend()
plt.grid()
plt.show()
```

#### 4.2 Faire des subplots

#### 4.2.1 Approche fonctionnelle

```python
# Graphique avec subplot (Approche fonctionnelle)
plt.subplot(2,2,1) # 2: nbre de lignes, 1: nbre de colonnes et 1: la position du graphique 
plt.plot(x,y,c="red",ls="-", label="exp")
plt.subplot(2,2,2)
plt.plot(x,y,c="blue",ls="-", label="exp")
plt.subplot(2,2,3)
plt.plot(x,y,c="green",ls="-", label="exp")
plt.subplot(2,2,4)
plt.plot(x,y,c="black",ls="-", label="exp")
plt.subplots_adjust(wspace=0.5)
```

#### 4.2.2  Approche orientée objet


```python
# Subplots
fig,axes=plt.subplots(2,1) # 2 lignes et 1 colonne
axes[0].plot(x,y,c="red")
axes[1].plot(x,y,c="blue")
```

```python
# Sauvegarder son graphique dans un fichier
plt.subplot(2,2,1) 
plt.plot(x,y,c="red",ls="-", label="exp")
plt.subplot(2,2,2)
plt.plot(x,y,c="blue",ls="-", label="exp")
plt.subplot(2,2,3)
plt.plot(x,y,c="green",ls="-", label="exp")
plt.subplot(2,2,4)
plt.plot(x,y,c="black",ls="-", label="exp")
plt.subplots_adjust(wspace=0.5)
plt.savefig("figure.png")
plt.savefig("figure.jpeg")
```
